```python
plt.figure(figsize=(8,8))
sns.countplot(x="Region", data=df, color='yellow')
plt.title("Number of Countries of the Regions")
plt.show()
```

```python
# Örnek olarak bazı ülkeleri seçelim
selected_countries = ['United States', 'United Kingdom', 'India', 'China', 'Russia']

# Seçili ülkelerin verilerini filtreleyelim
filtered_df = df[df['Country'].isin(selected_countries)]

# Çizgi grafiği oluşturalım
plt.figure(figsize=(10,6))
sns.lineplot(x='Country', y='GCI', data=filtered_df, marker='o')

plt.title('GCI (Global Cybersecurity Index) Comparison of Selected Countries')
plt.xticks(rotation=45)
plt.ylabel('Global Cybersecurity Index (GCI)')
plt.xlabel('Country')
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.boxplot(data=df, x="CEI", y="Region")
plt.title("Exposure to Cybercrime")
```

```python
plt.figure(figsize=(8,8))
sns.boxplot(data=df, x="NCSI", y="Region")
plt.title("Preparation to Manage Cyber ​​Incidents")
```

```python
plt.figure(figsize=(8,8))
sns.boxplot(data=df, x="DDL", y="Region")
plt.title("Average percentage at maximum values ​​of both CEI and GCI")
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='GCI', y='CEI', data=df, hue='Region', size='NCSI', sizes=(20, 200), palette='magma')

plt.title('Relationship Between GCI and CEI by Region')
plt.xlabel('Global Cybersecurity Index (GCI)')
plt.ylabel('Cybersecurity Exposure Index (CEI)')
plt.show()
```
